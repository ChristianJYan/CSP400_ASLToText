```python
import re
import os
import json
import numpy as np
import shutil
```

```python
with open('../Data/Videos/wlasl-complete/wlasl_class_list.txt') as f:
    vid_names = f.readlines()

for x in range(len(vid_names)):
    vid_names[x] = re.sub(r'[0-9]+\W', '', vid_names[x]) 
    vid_names[x] = vid_names[x][:-1]
    
print(vid_names)
```

```python
with open('../Data/Videos/wlasl-complete/WLASL_v0.3.json', 'r') as json_file:
    data = json.load(json_file)
    
# Filter the data
filtered_data = [item for item in data if item.get("gloss") in vid_names]

print(filtered_data[0]["instances"][0]["video_id"])
```

```python
#splits the videos into a dict of class name as key and videos as values
vid_dir = "../Data/Videos/wlasl-complete/videos"
id_dict = {}
for word in filtered_data:
    id_array = []
    label = ""
    for i in range(len(word["instances"])):
        #added .mp4 to the end for file moving
        vid_id = word["instances"][i]["video_id"]
        id_array.append(vid_id)
        #added file path before for file moving
        label = "../Data/Videos/wlasl-complete/classes/" + word['gloss']
    id_dict[label] = id_array
print(id_dict)
```
